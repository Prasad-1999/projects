```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path ="/content/file.csv"
```

```python
dataset_path = "/content/tested.csv"
```

```python
python_ds = pd.read_csv(dataset_path)
```

```python
python_ds
```

```python
python_ds.isnull().sum()
```

```python
python_ds.drop("Cabin",axis=1,inplace=True)
python_ds.drop("Ticket",axis=1,inplace=True)
python_ds.drop("Name",axis=1,inplace=True)
python_ds.drop("Fare",axis=1,inplace=True)

print(python_ds)
```

```python
python_ds.describe()
```

```python
med = python_ds["Age"].median()
med=int(med)
med
```

```python
python_ds["Age"]=python_ds["Age"].fillna(value=27)
python_ds["Age"] = python_ds["Age"].astype(int)
python_ds
```

```python
python_ds.isnull().sum()
```

```python
sns.heatmap(python_ds.corr(), annot=True, fmt='.0%')
```

```python
plt.hist(python_ds["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Embarked")
plt.show()
```

```python
sns.violinplot(x=python_ds["Survived"], y=python_ds["PassengerId"])

plt.xlabel("Survived")
plt.ylabel("PassengerId")
plt.title("Passengers survived rate")
plt.show()
     
```

```python
class_counts = python_ds["Survived"].value_counts()
plt.pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%')
plt.title("survived and not survived")
plt.show()
     
```

```python
sex_counts = python_ds["Sex"].value_counts()
plt.bar(sex_counts.index, sex_counts.values)
plt.xlabel("Sex")
plt.xticks([0, 1])
plt.ylabel("Sibsp")
plt.title("Passenger counts by sex")
plt.show()
```

```python
corr_matrix = python_ds.corr()
sns.heatmap(corr_matrix, annot=True, cmap="cool")
plt.title("Correlation matrix")
plt.show()
```

```python
plt.boxplot(python_ds["Parch"])
plt.xlabel("Parch")
plt.title("Parch rate")
plt.show()
```
